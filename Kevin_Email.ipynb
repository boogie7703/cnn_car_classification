# Functions

```python
def rmse(true_y,outcome):
    true_y = true_y.tolist()
    outcome = outcome.tolist()
    res = []
    for i in range(len(true_y)):
        dff = true_y[i] - outcome[i]
        res.append(dff**2)
    avg = sum(res)/len(res)
    return math.sqrt(avg)
```

```python
def cross_val(df,estimator, depth, split):
    df_copy = df
    num_sample = 2
    sample_size = int(len(df)/2)
    samples = []
    temp_model = RandomForestRegressor(n_estimators=estimator, max_depth=depth,min_samples_split=split)
    temp_rmse_list = []
    for i in range(num_sample):
        sample = df.sample(n=sample_size)
        samples.append(sample)
        df = df.drop(sample.index)
    for j in range(2):
        temp_df = df_copy.drop(samples[j].index)
        temp_X1 = temp_df.drop('spend',axis=1)
        temp_y1 = temp_df['spend']
        sample_X = samples[j].drop('spend',axis=1)
        sample_y = samples[j]['spend']
        temp_model.fit(temp_X1,temp_y1)
        temp_outcome = temp_model.predict(sample_X)
        temp_rmse = rmse(sample_y,temp_outcome)
        temp_rmse_list.append(temp_rmse)
    avg_rmse = sum(temp_rmse_list)/len(temp_rmse_list)
    return avg_rmse
```
